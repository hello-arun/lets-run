# Discrete Fourier transform (DFT)

The discrete Fourier transform transforms a sequence of N complex numbers 

$${x_n} := x_0, x_1, ...., x_{N-1}$$ 

into another seq of complex numbers,

$${X_k} := X_0, X_1, X_2, ...., X_N$$

which is defined by

$$X_k = \sum_{n=0}^{N-1} x_n. e^{\frac{-i2\pi}{N}kn} = \sum_{n=0}^{N-1} x_n. [ cos(\frac{2\pi}{N}kn) - i.sin(\frac{2\pi}{N}kn)]$$

```python
from math import sin,cos,pi
import numpy as np
import matplotlib.pyplot as plt
TWO_PI=2*pi
num_points=500
```

Let's generate some waveform and then try to find out its freq component

```python

X=np.linspace(0,6,num_points)*TWO_PI
Y=(np.cos(X)+np.cos(2*X)-np.cos(2.5*X))
plt.plot(X/TWO_PI,Y)
```

```python
f=1.2
plt.polar(X*f,Y)
```

Calculate centre of mass of this plot for different values of f

```python
samples=200
freqs=np.linspace(0,5,samples)
# center of mass
com=np.zeros((len(freqs),2)) 
f,(ax1,ax2)=plt.subplots(2)
for i,freq in enumerate(freqs):
    com_x=np.sum(Y*np.cos(X*freq))/num_points
    com_y=np.sum(Y*np.sin(X*freq))/num_points
    com[i,0]=com_x*2
    com[i,1]=com_y*2

ax1.plot(freqs,com[:,0],label="COM_x")
ax2.plot(freqs,com[:,1],label="COM_y")
ax1.legend()
ax2.legend()
```
