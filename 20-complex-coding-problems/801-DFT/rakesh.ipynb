```python
from math import sin,cos,pi
import numpy as np
import matplotlib.pyplot as plt
TWO_PI=2*pi
num_points=500
```

Let's generate some waveform and then try to find out its freq component

```python
data=np.loadtxt("data-rakesh.txt")
X=data[:,0]
Y=data[:,1]
num_points=len(X)
plt.plot(X/TWO_PI,Y)
```

```python
f=1.2
plt.polar(X*f,Y)
```

Calculate centre of mass of this plot for different values of f

```python
samples=200
freqs=np.linspace(0,10,samples)
# center of mass
com=np.zeros((len(freqs),2)) 
f,(ax1,ax2)=plt.subplots(2)
for i,freq in enumerate(freqs):
    com_x=np.sum(Y*np.cos(X*freq))/num_points
    com_y=np.sum(Y*np.sin(X*freq))/num_points
    com[i,0]=com_x*2
    com[i,1]=com_y*2

ax1.plot(freqs,com[:,0],label="COM_x")
ax2.plot(freqs,com[:,1],label="COM_y")
ax1.legend()
ax2.legend()

```
